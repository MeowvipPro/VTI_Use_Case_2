```python
# -------------------------
# Ask Gemini to analyze
# -------------------------
prompt = f"""
You are a financial analyst. Analyze the following stock historical data.
Return:
- Trends
- Support / resistance levels
- Volatility summary
- Outlier days
- Buy/Sell sentiment (based on TA signals)

Data:
{data_json}

PLS MAKE FINAL DECICIONS BUY OR NOT
"""

config = GenerateContentConfig(
    temperature=0.2,
    max_output_tokens=2048
)

response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=prompt,
    config=config
)

print(response.text)
```

```python
df_stock["ai_decision"] = None      # new column
df_stock["ai_reason"] = None        # NEW: reason column

# -----------------------------------------------------
# Function: ask Gemini for decision for ONE DAY
# -----------------------------------------------------
def get_gemini_decision(history_df):
    data_json = history_df.to_dict(orient="records")

    prompt = f"""
You are a financial analyst AI.
Analyze the stock data and answer ONLY in this format:

DECISION: BUY or SELL or HOLD
REASON: <one short line>

Data:
{data_json}
"""

    cfg = GenerateContentConfig(
        temperature=0.1,
        max_output_tokens=200
    )

    resp = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=prompt,
        config=cfg
    )

    return resp.text.strip()


# -----------------------------------------------------
# Loop day-by-day
# -----------------------------------------------------
WINDOW = 1  # number of days Gemini should analyze

for i in range(len(df_stock)):
    if i < WINDOW:
        continue  # wait until enough data

    history = df_stock.iloc[i-WINDOW:i+1]

    result = get_gemini_decision(history)

    # -----------------------------------------------------
    # Parse DECISION and REASON from Gemini response
    # -----------------------------------------------------
    lines = result.split("\n")
    decision = "HOLD"
    reason = ""

    for line in lines:
        line_upper = line.upper()

        if "DECISION:" in line_upper:
            if "BUY" in line_upper:
                decision = "BUY"
            elif "SELL" in line_upper:
                decision = "SELL"
            else:
                decision = "HOLD"

        if "REASON:" in line_upper:
            reason = line.split(":", 1)[1].strip()

    df_stock.at[i, "ai_decision"] = decision
    df_stock.at[i, "ai_reason"] = reason

    print(f"Day {i} → {decision} | Reason: {reason}")
```

```python
import pyodbc

# --- Connect ---
conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=DESKTOP-OQLFA92\\SQLEXPRESS;"
    "DATABASE=vti_dataset;"
    "Trusted_Connection=yes;"
    "TrustServerCertificate=yes;"
)
cursor = conn.cursor()

# --- Map pandas dtype -> SQL Server type ---
dtype_map = lambda dt: "INT" if pd.api.types.is_integer_dtype(dt) else \
                       "FLOAT" if pd.api.types.is_float_dtype(dt) else \
                       "BIT" if pd.api.types.is_bool_dtype(dt) else \
                       "DATETIME" if pd.api.types.is_datetime64_any_dtype(dt) else \
                       "VARCHAR(255)"

table_name = "use_case_2_ai_stock" #

# --- Drop + Create table ---
cursor.execute(f"IF OBJECT_ID('dbo.{table_name}', 'U') IS NOT NULL DROP TABLE dbo.{table_name}")
cols = ", ".join([f"[{c}] {dtype_map(df_stock[c].dtype)}" for c in df_stock.columns]) #
cursor.execute(f"CREATE TABLE dbo.{table_name} ({cols})")

# --- Insert rows ---
cursor.fast_executemany = True
cursor.executemany(f"INSERT INTO dbo.{table_name} VALUES ({','.join(['?']*len(df_stock.columns))})", df_stock.values.tolist()) #
conn.commit()
```
